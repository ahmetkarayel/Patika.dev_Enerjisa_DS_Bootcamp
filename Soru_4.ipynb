## Test Duyarlılıklarının Hesaplanması

Herhangi bir hastalığın teşhisi için yeni bir test geliştirildiğini farz edelim. Bu hastalığın insanlarda görülme sıklığının 5/1000 olduğunu farz edelim, bu test için ön bilgimiz, testin hasta olanlar üzerinde denendiğinde 95% pozitif sonuç verdiği olsun. Bu test güvenilir midir? 95% yüksek  bir oran  olduğu için  testin güvenli  olduğu yolunda  hissiyata sahip olabiliriz.

Bayes Teoremine göre;

a = testin uygulandığı kişide pozitif sonuç vermesi (teste göre "kişi hasta")   
b = kişinin hasta olması (teste göre değil, gerçekten hasta)  

Böylece;   
a' = testin uygulandığı kişide negatif sonuç vermesi,   
b' = kişinin hasta olmamasıdır.   

Bu durumda başta verilen bilgileri kullanırsak;   

>> p[a|b] = p[a'|b'] = 0.95 (kişi hasta (b) ve test uygulanmış(a))

>> p[b] = 0.005 (bir kişinin hasta olma olasılığı, ya da bu hastalığa rastlanma sıklığı)

*Sonuç*  
* Bu test pozitif sonuç verdiğinde aslında o kişinin hasta olma olasığı 8.7% dir. Bu değere göre testin %91,3 oranında yanlış pozitif sonuç vereceği ortaya çıkmaktadır. Bu durumda bu test güvenilmezdir diyebiliriz.

## Spam E-mail Tespit

Hesabımdaki spam iletilerin %80'inde "teklif" kelimesinin geçtiğini varsayalım. Ayrıca spam olmayan iletilerinde %10'unda "teklif" kelimesinin geçtiğini varsayalım. Gelen e-maillerin %30'u dolandırıcılık olarak kabul edilirse ve "teklif" kelimesi içeren yeni bir mail alırsam, bunun spam olma olasılığı nedir?

Şimdi, 100 e-email aldığımı varsayıyorum. Tüm e-maildeki spam yüzdesi %30'dur. Yani, 100 e-mail'de arasında 30 istenmeyen e-mail ve 70 istenen e-email var. Spam e-postalarda "teklif" kelimesi geçme oranı %80 idi. 30 spam e-mailin %80'i 24'tür.  

Şimdi, 100 e-mailin 30'unun spam ve bunların 24'ünün "teklif" kelimesini içerdiğini ve kalan 6 tanesinin "teklif" kelimesini içermediğini biliyorum.

Spam olmayan 70 e-mail'in %10'unda "teklif" kelimesinin geçtiğini bildiğime göre bu e-maillerden 7 tanesi "teklif" kelimesini içerirken 63 tanesi içermez.  

resim_2021-10-05_141732.png




"Teklif" kelimesini içeren ve spam olma olasılığını arıyoruz.  

A = spam  
B = "teklif" kelimesi içeren mail

resim_2021-10-05_142233.png


* P(teklif | spam ) = 0.8 ( "teklif" kelimesi içeren spam mail. Soruda verilmişti.)

* P(spam ) = 0.3 ( Gelen maillerin %30'u spam. Soruda verilmişti.)

* P(teklif) = 0.3 * 0.8 + 0.7 * 0.1 = 0.31

>> **P(spam | teklif) = (0.8 * 0.3) / 0.31 = 0.774** olarak buluruz. 




## Kaza Problemi

Gece vakti bir taksi, bir yayaya çarparak kazaya karışmıştır. Şehirde "Green" ve "Blue" adlı iki farklı taksi şirketi faaliyet göstermektedir. Bu kaza hakkında bilinen bilgiler şu şekilde;  

* Şehirdeki taksilerin %85'i yeşil, %15'i mavidir.  
* Bir görgü tanığı taksinin mavi olduğunu iddia etmektedir. Mahkeme, kaza gecesi var olan aynı koşullar altında tanığın güvenirliğini test etmiş ve tanığın her iki rengin her birini %80 oranında doğru bir şekilde tanımladığı ve %20 oranında başarısız olduğu sonucuna varmıştır.
* Bu tanığın onu mavi olarak tanımladığını düşünürsek, kazaya karışan taksinin yeşil yerine mavi olma olasılığı nedir?
---

* Problemimizde taksiyi mavi olarak düşünürsek taksinin mavi olma olasılığı ile ilgileniyoruz.

>> P(Taksi Mavi | Mavi) denklemi için;
* `P(Taksi Mavi) = 0.15`
* Taksinin gerçekten mavi iken mavi olarak görülme olasılığı. Bu bize  mahkemenin güvenilirlik testinde verilmişti. Tanığın taksinin rengini %80'inde doğru şekilde tanımladığını biliyoruz.  
`P(Mavi Olarak Görüldü | Taksi Mavi) = 0.8`
* Taksinin mavi görünme olasılığı. Bu iki farklı şekilde olabilir: Tanık mavi bir taksiyi doğru bir şekilde tanımlamıştı veya yeşil bir taksiyi yanlış bir şekilde mavi olarak tanımladı.  
`P (Mavi Olarak Görülme Olasılığı) = (0.80) * (0.15) + (0.20) * (0.85) = 0.29`

* Şimdi koşulu hesaplarsak:  
> `P(Taksi Mavi | Mavi Olarak Görülme) = (0.15) * (0.80) / (0.29) = 0.41` olarak bulunur.
